```python
import urllib.request
import json
import pandas as pd
```

```python
pidvar = ["NTN001"]
yearvar = ["2017", "2018"]

# 기본 URL
base_url = "http://www.recycling-info.or.kr/sds/JsonApi.do?PID={}&YEAR={}&USRID=eunjungcho425&KEY=U2B5HB0ZUFOHR40T5T0QSV8L2DBQCZPGQ734T1QIMENPR"

# 데이터를 저장할 리스트
data_list = []
```

```python
for pid in pidvar:
    for year in yearvar:
        url = base_url.format(pid, year)

        response = urllib.request.urlopen(url)
        json_str = response.read().decode("utf-8")


        data = json.loads(json_str)


        if 'data' in data:
            for entry in data['data']:
                entry['PID'] = pid
                entry['YEAR'] = year
                data_list.append(entry)
```

```python
data_list
```

```python
df = pd.json_normalize(data_list)
#df.tail()
```

```python
df = df.drop(columns=['result', 'dataHeader', 'data', 'searchOption'], errors='ignore')
df.head()
```
