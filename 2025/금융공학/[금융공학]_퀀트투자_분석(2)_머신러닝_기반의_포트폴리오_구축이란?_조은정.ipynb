```python
import pyreadr
import pandas as pd
import os

# 파일 저장된 위치 확인하기
file_path = "/content/drive/MyDrive/EJ_Data Analysis/금융공학/data_ml.RData" # 이 부분은 저장위치에 따라 바꾸면 된다.
if not os.path.exists(file_path):
  print(f" 파일을 찾을 수 없다. : {file_path}")
else:
  # Rdata 파일 불러오기
  try:
    result = pyreadr.read_r(file_path)
    # 1객체만 있다면 데이터프레임으로 변환
    if len(result) == 1:
      key = list(result.keys())[0]
      df = result[key]
      # 확인용
      print(df.head())
      print(df.shape)

    else :
      print(f"1개의 객체보다 더 많은 파일이 존재 : {file_path}")
      for key in result.keys():
        print(f"{key} : {result[key].shape}")
  except Exception as e:
      print(f"에러 발생한 곳: {e}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib

df_median = []  # 임시 데이터프레임을 저장할 빈 리스트 생성
df_2 = []
df_median = df[['date', 'Mkt_Cap_12M_Usd']].groupby(['date']).median().reset_index()  # 'date'별로 중앙값 계산
df_median.rename(columns={'Mkt_Cap_12M_Usd': 'cap_median'}, inplace=True)  # 열 이름을 명확히 변경
df_2 = pd.merge(df[["date", 'Mkt_Cap_12M_Usd', 'R1M_Usd']], df_median, how='left', on=['date'])  # 중앙값 데이터와 병합
df_2 = df_2.groupby([pd.to_datetime(df_2['date']).dt.year, np.where(df_2['Mkt_Cap_12M_Usd'] > df_2['cap_median'], '대형주', '소형주')])['R1M_Usd'].mean().reset_index()  # 'year'와 'cap' 기준으로 그룹화 및 평균 계산
df_2.rename(columns={'level_1': '시가총액분류'}, inplace=True)  # 열 이름을 명확히 변경
df_2.pivot(index='date', columns='시가총액분류', values='R1M_Usd').plot.bar(figsize=(10, 6))  # 피벗 테이블 생성 후 막대 그래프 그리기
plt.ylabel('평균 수익률')  # y축 레이블 설정
plt.xlabel('연도')  # x축 레이블 설정
df_median = []  # 임시 데이터프레임 삭제하여 메모리 절약
df_2 = []  # 임시 데이터프레임 삭제하여 메모리 절약
```

```python
import urllib.request
import zipfile
import pandas as pd

# 설정
min_date = '1963-07-31'
max_date = '2020-03-28'
ff_url = "https://mba.tuck.dartmouth.edu/pages/faculty/ken.french/ftp/F-F_Research_Data_5_Factors_2x3_CSV.zip"

# 파일 다운로드
urllib.request.urlretrieve(ff_url, 'factors.zip')  # 'factors.zip'으로 다운로드

# 압축 해제
with zipfile.ZipFile('factors.zip', 'r') as zip_ref:
    zip_ref.extractall()  # 현재 디렉토리에 압축 해제

# 압축 해제된 파일 읽기
df_ff = pd.read_csv('F-F_Research_Data_5_Factors_2x3.csv', header=2, sep=',', quotechar='"')

```

```python
# 기준 인덱스 찾기
split_index = df_ff[df_ff['date'] == ' Annual Factors: January-December '].index[0]

# 데이터프레임 분리
df_ff_ym = df_ff.iloc[:split_index]  # 기준값 위쪽 데이터
df_ff_yy = df_ff.iloc[split_index + 1:]  # 기준값 아래쪽 데이터
```

```python
df_ff_ym['date'] = pd.to_datetime(df_ff_ym['date'], format='%Y%m')
```

```python
df_ff_ym.iloc[:, 1:] = df_ff_ym.iloc[:, 1:].astype(float)
```

```python
# 데이터 준비
x = df_ff_ym['date']  # x축: date 열
y_columns = df_ff_ym.columns[1:]  # 나머지 열들 (y축)

# 그래프 그리기
plt.figure(figsize=(20, 6))  # 그래프 크기 설정

for col in y_columns:
    plt.plot(x, df_ff_ym[col], label=col)  # 각 열을 라인 그래프로 추가

# 그래프 꾸미기
plt.title('Line Graph of df_ff_ym', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend(title='Columns', fontsize=10)  # 범례 추가
plt.grid(True)  # 그리드 추가
plt.tight_layout()  # 레이아웃 조정

# 그래프 표시
plt.show()
```

```python
df_ff_yy.iloc[:, 1:] = df_ff_yy.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
```

```python
# date 컬럼을 제외한 나머지 컬럼 변환
for col in df_ff_yy.columns:
    if col != 'date':  # date 컬럼 제외
        df_ff_yy[col] = pd.to_numeric(df_ff_yy[col], errors='coerce')  # 변환 불가능한 값은 NaN으로 처리

# 변환할 수 없는 값이 있는 행 삭제
df_ff_yy = df_ff_yy.dropna()  # NaN이 포함된 행 제거
```

```python
# date 컬럼을 int로 변환 (연도를 나타내기 위해)
df_ff_yy['date'] = pd.to_numeric(df_ff_yy['date'], errors='coerce')  # 변환 불가능한 값은 NaN으로 처리
df_ff_yy = df_ff_yy.dropna(subset=['date'])  # date 컬럼에서 NaN 제거
df_ff_yy['date'] = df_ff_yy['date'].astype(int)  # int로 변환

# 데이터 준비
x = df_ff_yy['date']  # x축: date 열
y_columns = df_ff_yy.columns[1:]  # 나머지 열들 (y축)

# 그래프 그리기
plt.figure(figsize=(15, 6))  # 그래프 크기 설정 (가로 15, 세로 6)

for col in y_columns:
    plt.plot(x, df_ff_yy[col], label=col)  # 각 열을 라인 그래프로 추가

# 그래프 꾸미기
plt.title('일반적인 이상 현상들의 평균 수익률(1963~2020)', fontsize=16)
plt.xlabel('연도', fontsize=12)
plt.ylabel('평균수익률', fontsize=12)
plt.legend(title='팩터들', fontsize=10)  # 범례 추가
plt.grid(True)  # 그리드 추가
plt.tight_layout()  # 레이아웃 조정

# 그래프 표시
plt.show()
```
