```python
# 경로 설정
directory = '/content/drive/MyDrive/DB/소상공인시장진흥공단_상가(상권)정보_20230331'
file_prefix = '소상공인시장진흥공단_상가(상권)정보'

# CSV 파일 불러와서 데이터프레임으로 합치기
dataframes = []
for file in os.listdir(directory):
    if file.startswith(file_prefix) and file.endswith('.csv'):
        file_path = os.path.join(directory, file)
        df = pd.read_csv(file_path)
        dataframes.append(df)

merged_df = pd.concat(dataframes)

# 결과 확인
print(merged_df.head())
```

```python
dataset = merged_df[['상호명','지점명',
              '상권업종대분류명', '상권업종중분류명', '상권업종중분류코드',
              '시도명', '시군구명', '행정동명',
              '위도', '경도']]

dataset.head()
```

```python
df_coffee = dataset[(dataset['상권업종중분류코드']=='I212')&(dataset['시도명']=='서울특별시')]
df_coffee.index = range(len(df_coffee))
print('서울시 내 커피 전문점 점포 수 :', len(df_coffee))
df_coffee.head()
```
